```python
# import the required libraries
import numpy as np
import pandas as pd
```

```python
# read the dataset
df = pd.read_csv("D:\mtech\ML\Asssignments\module\Adult.csv")
```

```python
# Drop education as education and education number will convey same meaning and with this we can delete duplicated data
#df.drop(['education'],axis=1)
```

```python
df["fnlwgt"].value_counts
```

```python
# remove fnlwgt as it the range of values are very high
df = df.drop(["education","fnlwgt"],axis=1)
```

```python
# replace '?'' with 'NaN'
df.replace('?',np.NaN,inplace=True)
```

```python
df
```

```python
# repalce NaN with forward fill
df.fillna(method='ffill',inplace=True)
```

```python
df.head()
```

```python
# Need to fit the data Otherwise while fitting training and testing data we can't compare
from sklearn.preprocessing import LabelEncoder


le = LabelEncoder()
df['workclass'] = le.fit_transform(df['workclass'].astype(str))
df['marital.status'] = le.fit_transform(df['marital.status'])
df['occupation'] = le.fit_transform(df['occupation'].astype(str))
df['relationship'] = le.fit_transform(df['relationship'])
df['race'] = le.fit_transform(df['race'])
df['sex'] = le.fit_transform(df['sex'])
df['native.country'] = le.fit_transform(df['native.country'])
df['income'] = le.fit_transform(df['income'])

df.head()
```

```python
# replace x with all columns except income
x = df.drop(['income'],axis=1)
```

```python
# replace y with income column
y = df['income']
```

```python
#split the data into train and test data
# 30% test data and 70% training data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
gb=GaussianNB()
gb.fit(x_train,y_train)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
y_pred = gb.predict(x_test)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred)*100)
```

```python
from sklearn.naive_bayes import BernoulliNB
bnb = BernoulliNB()
bnb.fit(x_train,y_train)
```

```python
y_pred = bnb.predict(x_test)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred)*100)
```

```python
from sklearn.naive_bayes import MultinomialNB
MnB = MultinomialNB()
MnB.fit(x_train,y_train)
```

```python
y_pred = MnB.predict(x_test)
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred)*100)
```
